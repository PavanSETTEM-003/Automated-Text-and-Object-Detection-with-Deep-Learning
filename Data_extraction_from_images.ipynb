## Text extraction

```python
image_folder_path = r"/content/drive/MyDrive/DKNSB/images"    # drive folder location 
img_folder = sorted(os.listdir(image_folder_path))            # using os all the files in the folder path were obtained
```

```python
extracted_data = []                           ## data extracted using library
cropped = []                                  ## small piece of image was cropped and the text extracted from it was stored over here
features = ["Device Name: ","REF ","LOT: "]   ## features which need to be extracted
values_found = []                             ## Final values obtained were stored in it 
```

```python
## looping over each and every image in the folder path
for i in img_folder:
  img = cv2.imread(image_folder_path+ '/' + i,0)
  extractedInformation = pytesseract.image_to_string(img)        ## supports jpg format
  extracted_data.append(extractedInformation)
  

  cropped_image = img[0:500,1000:1400 ] 
  extracted = pytesseract.image_to_string(cropped_image)
  cropped.append(extracted)

    
```

```python
def feature_extraction(extracted_data):
  found = []
  ref = extracted_data
  ref = ref.replace("\n\n","\n")
  pre_process = ref.split("\n")
  found.append(pre_process[0])
  c = 0
  for i in pre_process[1:4]:
    feature = features[c]
    found.append(i.replace(feature,""))
    c+=1
  return found
```

```python
for i in extracted_data:                       ## extracting features values from the text_extraceted
  values_found.append(feature_extraction(i))

for i in range(len(cropped)):                    ## extracting quantity
  temp = re.findall(r'\d+',cropped[i])
  values_found[i].append(temp[0])
```

## Symbols extraction

```python
5 0.46601 0.596154 0.113528 0.0948718 0.826414
CLASS // X-COOR // Y-COOR // WIDTH // HEIGHT // PROBABILITY
```

```python
labels_folder_path = r"/content/drive/MyDrive/DKNSB/results/labels_result"
labels_folder = sorted(os.listdir(labels_folder_path))    
```

```python
symbols_order = []         ## final result of each image stored over here
label_values=[]            ## all the values in the label.txt file were appended here
```

```python
for i in labels_folder:
  f = open(labels_folder_path+"/"+i, "r")
  label_values.append(f.read())
```

```python
def second_level(x):
  copy = x
  ref = x[0][1]
  for i in range(len(x)):
    if(abs(float(x[i][1])-float(ref))>=0.1):
      copy.append(x[i])
      del copy[i]
  return copy

```

```python
for x in label_values:
  dict = {}
  found = []
  for i in x.split("\n")[0:-1]:
    n = list(i.split(" "))
    if(n[1] not in dict):
      dict[n[1]] = [n[0],n[2]]
  print(dict)
  for keys in sorted(dict):
    found.append(dict[keys])
  found = second_level(found)   ## sorting based on y axis at second level function

  
  ref = ""
  for i in found:
    if(i[0]!="3"):
      ref += str(int(i[0])+1)   ## as indexing start with zero i have added once too each prediction of class
    else:
      ref = "4"+ref
  symbols_order.append(ref) 
```
